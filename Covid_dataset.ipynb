```python
from numpy import datetime64, float64, int64
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.impute import SimpleImputer
```

```python
# Ensure column names are strings
covid_file.columns = covid_file.columns.astype(str)
# Define categorical pipeline
catPipe = Pipeline([
    ("impute", SimpleImputer(strategy="most_frequent")),
])
# Define quantitative pipeline
quantPipe = Pipeline([
    ("impute", SimpleImputer(strategy="median"))
])

# Combine pipelines in a ColumnTransformer
fullPipe = ColumnTransformer([
    ("quant", quantPipe, make_column_selector(dtype_include=[float64, int64])),
    ("cat", catPipe, make_column_selector(dtype_include=object))
])

# Fit and transform the data
piped = fullPipe.fit_transform(covid_file)

# Create a new DataFrame with transformed data
covid_file_transformed = pd.DataFrame(piped, columns=fullPipe.get_feature_names_out())
```
